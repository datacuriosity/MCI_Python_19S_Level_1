```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np 
import pandas as pd
```

```python
x = np.linspace(-10, 10, 1000)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(10, 6))

plt.plot(x, sin, label='Sin func')
plt.plot(x, cos, label='Cos func')

plt.title('Sin function')

plt.xlabel('x axis label')
plt.ylabel('Y axis label')

plt.legend(loc='upper left');
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips.total_bill, tips.tip, c='red', marker = 's')
plt.xlabel('Total bill')
plt.ylabel('Tip');
```

```python
tips.plot(kind='scatter', x='total_bill', y='tip');
```

```python
plt.hist(tips.total_bill, color='blue', bins=50);
```

```python
tips.plot(kind ='hist',y='total_bill', bins = 50);
```

```python
data = tips.groupby('day').total_bill.sum().sort_values(ascending=False)
```

```python
plt.pie(data.values, labels=data.index, autopct='%.2f%%', 
        normalize=True, radius=2, explode = (0,0,0,0.2));
```

```python
plt.figure(figsize=(10,4))
plt.boxplot(tips.tip, vert=False);
```

```python
plt.subplot(1,2,1)
plt.bar(data.index,data.values)
plt.subplot(1,2,2)
plt.pie(data.values, labels=data.index, autopct='%.2f%%', 
        normalize=True, radius=1.3, explode = (0,0,0,0.2));
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,10))
axes[0,0].bar(data.index,data.values)
tips.plot(kind ='hist',y='total_bill', bins = 50, ax = axes[0,1])
axes[1,0].pie(data.values, labels=data.index, autopct='%.2f%%', 
        normalize=True, radius=1.3, explode = (0,0,0,0.2));
```
