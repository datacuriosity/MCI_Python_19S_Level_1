```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#step 1: read and display number of rows

```python
data=pd.read_csv('C://Users//HP//Desktop//PYTHON PROJECT EXP//smokers.csv')
print(len(data))
```

#step 2: Find and display the average number of surveys conducted per state


```python
print(data['State'].nunique())
print(len(data)/data['State'].nunique())
```

#Step 3: Among the surveys in the data-set, display the minimum and maximum smoking rate.

```python
print(data[data['Value']==data['Value'].min()])
print(data[data['Value']==data['Value'].max()])
```

#Step 4:  Find average smoker rate per state anddisplay the states with minimum and maximum average


```python
by_State=data.groupby(by='State')
mean_by_State=by_State['Value'].mean().reset_index()
print(mean_by_State)

min_by_State=mean_by_State[mean_by_State['Value']==mean_by_State['Value'].min()]
print(min_by_State)

max_by_State=mean_by_State[mean_by_State['Value']==mean_by_State['Value'].max()]
print(max_by_State)
```

#Step 5 Same as Step 4 but display the top 10 states instead of min and max.

```python
print(mean_by_State.sort_values('Value',ascending=False)[0:10])
```

#Step 6: Get state and year from user and display

#Step 7: Get state from user as input (using
input()) and display the average smoker rate year by
year. By looking at the most recent 2 years, display if
the cigarette use is on decline or rise

```python
by_Year_State=data.groupby(by=['Year','State'])
mean_by_Year_State=by_Year_State.mean().reset_index()

plot_prompt=str(input('enter state to plot:'))
mean_by_Year_State[mean_by_Year_State['State']==plot_prompt].plot(kind='line',x='Year',y='Value')
plt.show()

latest=mean_by_Year_State[mean_by_Year_State['State']==plot_prompt]['Year'].max()
latest_m2=latest-2

latest_value=float(mean_by_Year_State[mean_by_Year_State['State']==plot_prompt][mean_by_Year_State['Year']==latest]['Value'])
latest_m2_value=float(mean_by_Year_State[mean_by_Year_State['State']==plot_prompt][mean_by_Year_State['Year']==latest_m2]['Value'])
print(latest_value>latest_m2_value)

if latest_value<latest_m2_value:
    print('the trend is decreasing')
elif latest_value>latest_m2_value:
    print('the trend is increasing')
else:
    print('there is no change in trend')
```
