## Bài 1.

Viết một chương trình tính giai thừa của một số nguyên dương n. Với n được nhập từ bàn phím. Ví dụ, n = 8 thì kết quả đầu ra phải là `1*2*3*4*5*6*7*8` = 40320.

```python
def factorial(n):
    result=1
    for i in range(1,n+1):
        result=i*result
    return result
```

```python
# Testing
n = 5
assert factorial(n) == 120, "Wrong answer, please calculate again"
```

## Bài 2.

Viết chương trình tính tổng của các chữ số của môt số nguyên n trong Python. Số nguyên dương n được nhập từ bàn phím. Với n = 1234, tổng các chữ số: 1 + 2 + 3 + 4 = 10

```python
def sum_all(number):
    result=sum(int(i) for i in str(number))
    return result
```

```python
# Testing
number = 34103
assert sum_all(number) == 11, "Wrong answer, please calculate again"
```

## Bài 3.

Tìm giá trị lớn nhất của một list hoặc tuple nhập vào

```python
def maximum(li):
    m=0
    for i in li:
        if i>m:
            m=i
        else:
            m=m
    return m
```

```python
# Testing
number = [1, 2, 6, 8, 10]
assert maximum(number) == 10, "Wrong answer, please calculate again"
```

## Bài 5.

Định nghĩa một hàm có input là 2 chuỗi và trả về chuỗi có độ dài lớn hơn. Nếu 2 chuỗi có chiều dài như nhau thì trả tất cả các chuỗi dưới dạng tuple

```python
def greater_string(str_1, str_2):
    result=""
    if len(str_1)>len(str_2):
        result=str_1
    elif len(str_1)<len(str_2):
        result=str_2
    else:
        result=(str_1,str_2)
    return  result
```

```python
# Test 1
str_1 = "My name is"
str_2 = "My university"

assert len(greater_string(str_1, str_2)) == len(str_2), "Wrong answer, please code again"


#Test 2

str_1 = "My name is"
str_2 = "My name ar"

assert greater_string(str_1, str_2) == (str_1, str_2), "Wrong answer, please code again"
```

## Bài 6.

Loại bỏ các giá trị trùng từ chuỗi nhập vào, nhưng vẫn giữ thứ tự.

Ví dụ: aaabbbbcccddd1111 --> abcd1

```python
def remove_duplicated(str_1):
    my_list = []
    my_string = ''
    for i in str_1:
        if i not in my_list:
            my_list.append(i)
    for i in my_list:
        my_string = my_string + i
    return  my_string
```

```python
# Testing
str_1 = "cccccddddeeeeffff22222"

assert remove_duplicated(str_1) == "cdef2"
```

Viết một chương trình có 2 chữ số X, Y nhận giá trị đầu vào và tạo ra một mảng 2 chiều. Giá trị phần tử trong hàng thứ i và cột thứ j của mảng phải là `i*j`

Lưu ý: 
- i = 0, 1, 2, ..., X-1
- j = 0, 1, 2, ..., Y-1      

```python
def create_2D_array(x, y):
    list = []
    for y in range(5):
        sub_list = []
        for x in range(5):
            sub_list.append(y * x)
        list.append(sub_list)
    return list
```

## Bài 8.
Viết chương trình để xuất một số ngẫu nhiên, chia hết cho 5 và 7, từ 0 đến 200 (gồm cả 0 và 200), sử dụng module random và list comprehension

```python
import random
def division_5_7(random_seed = 4):
    random.seed(random_seed)
    x=random.choice([i for i in range(200) if i%5==0 and i%7==0])
    return x
```

```python
random_seed = 4
assert division_5_7(random_seed) == 35, "Wrong answer, please code again"
```
