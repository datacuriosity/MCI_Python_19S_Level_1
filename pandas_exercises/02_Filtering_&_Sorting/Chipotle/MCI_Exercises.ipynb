### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
path = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(path, sep='\t')
```

```python
chipo.head(10)
```

### Step 4. How many products cost more than $10.00?

```python
x = '$10.98'
change = lambda x : float(x[1:])
```

```python
chipo['item_price'] = chipo.item_price.apply(change)
```

```python
chipo.query('item_price > 10').item_name.nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.groupby('item_name').item_price.mean().to_frame()
```

### Step 6. Sort by the name of the item

```python
chipo.groupby('item_name').item_price.mean().to_frame().sort_index()
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# most_exp = chipo.groupby('item_name').item_price.max().sort_values(ascending=False).head(1).index[0]
most_exp = chipo.groupby('item_name').item_price.max().nsmallest(5).index
```

```python
most_exp
```

```python
quantities = []
for item in most_exp:
    quantity = chipo.query("item_name == @item").quantity.sum()
    quantities.append(quantity)
```

```python
quantities
```

```python
pd.DataFrame({'item': most_exp, 'quantities': quantities})
```

```python
chipo.query("item_name == @most_exp").quantity.sum() # hard-code
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo.query("item_name == 'Veggie Salad Bowl'").order_id.nunique()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo.query("item_name == 'Canned Soda' & quantity > 1").order_id.nunique()
```
